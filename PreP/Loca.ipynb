```python
def createDatasets():
    indicators= ['tasmax',
 'xs',
 'tasavg',
 'dry',
 'xpr',
 'tasmin',
 'cum_pr',
 'cdds',
 'hdds',
 'ffs']
    periods=['decadal','30_y']
    scenarios=['rcp45','rcp85']
    datasets=[]
    for indicator in indicators:
        for scenario in scenarios:
            for period in periods:
                body={
                "connectorType":"rest",
                "provider":"loca",
                "application":[
                 "prep"
                ],
                "name":f"loca_{indicator}_{scenario}_{period}",
                "tableName": f"loca_{indicator}/{scenario}_{period}",
                "environment": "preproduction"
                    }
                r=requests.post(apiBaseUrl,json=body, headers=Headers)
                if r.status_code==200:
                    datasets.append(f'loca_{indicator}_{scenario}_{period}')
                else:
                    print(r.status_code)
                    print(r.text)
    return  datasets 
    
```

```python
def getDatasets(datasets):
    """
    Gets the list of dataset names and retrieves the name and the id of each dataset.
    """
    nDatasets=[]
    for name in datasets:
        params={
        'slug':name,
        'provider':Provider,
        'env':'preproduction,production'
        }
        r=requests.get(apiBaseUrl, params=params)
        ind=name.split('_')[1]
        if r.status_code==200 and r.json()['data']:
            nDatasets.append({'name':name,'datasetId':r.json()['data'][0]['id'], 'indicator': 'cum_pr' if ind =='cum' else ind, 'ramp':dict_key[ind], 'tags': tags_key[ind]})
    return nDatasets
```

### tags

```python
def checker(datasetId):
    apiUrl='{0}{1}/vocabulary?app=prep'.format(apiBaseUrl,datasetId)
    r=requests.get(apiUrl,headers=Headers)
    if r.status_code==200:
        print('****')
        print(apiUrl)
        print('****')
        
        if r.json()['data']:
            return True
        else:
            return False
    else:
        r.raise_for_status()
    
def post(tags, datasetId):
    apiUrl='{0}{1}/vocabulary'.format(apiBaseUrl,datasetId)
    body={
        'knowledge_graph': {
            'tags': tags, 
            'application': "prep"}
    }
    r=requests.post(apiUrl,json=body, headers=Headers)
    print('//////////////////////////')
    print(r.text)
    print('//////////////////////////')
    return r.status_code

def patch(tags, datasetId):
    apiUrl='{0}{1}/vocabulary/knowledge_graph'.format(apiBaseUrl,datasetId)
    body={
        'tags': tags, 
        'application': "prep"
    }
    
    r=requests.patch(apiUrl,json=body, headers=Headers)
   
    print(r.text)
    
    return r.status_code
    
def requestHandler(dataset):
    n=dataset['name'].split('_')[1]
    tags=tags_key[n]
    did=dataset['datasetId']
    checks = checker(dataset['name'])
    print(did)
    print(checks)
    print('//////////////////////////')
    if checks==False:
        sd = post(tags, did)
    elif checks==True:
        sd = patch(tags, did)
        
    print(sd)
```

## Layer

```python
def layerGenerator(datasets):
    #color ramp generator 
    Ramps={}
    for key, value in breaks_colized.items():
        print(key)
        ramp=[{'color':color,'stop':value[idx]} for idx, color in enumerate(colorRamps[key])]
        Ramps[key]={'colorized':ramp}
    for key, value in breaks_legend.items():
        ramp=[{'color':color,'value':value[idx]} for idx,color in enumerate(colorRamps[key])]
        Ramps[key]['legend']=ramp
        Ramps[key]['legend'].insert(0,{                             
       "color": "transparent",     
       "value": "No Data"          
         })
        print(Ramps[key]['legend'])
    
    #layer generator
    datasetsLayers=[]
    for dataset in datasets:
        nas=dataset['name'].split('_')
        if 'rcp45' in nas:
            t = nas.index('rcp45')
            s = len(nas[t:])
        else:
            t= nas.index('rcp85')
            s = len(nas[t:])

        if s == 2:
            name='{0}'.format(N_key[nas[1]],nas[t],tem_key['decadal'])
            scenario=nas[t]
            temp_resolution='decadal'
            
        else:
            name='{0}'.format(N_key[nas[1]],nas[t],tem_key['y'].format(30))
            scenario=nas[t]
            temp_resolution='30_y'
        print(units_key)
        baseLayer={
        'name':name,
        'provider': Provider,
        'application': ['prep'],
        "env": "production",
        "default": True,
        'layerConfig':{
            'colorRamp':Ramps[dataset['ramp']]['colorized'],
            "indicator": dataset['indicator'],
            "noData": -1,
            "type": "nexgddp",
            "scenario":scenario,
            "temp_resolution":temp_resolution,
            "bbox": [
                  -129.55078125,
                  22.268764039073968,
                  -66.26953125,
                  50.17689812200107
                ]
            },
        'legendConfig':{
                'type': 'gradient',
                'unit': units_key[nas[1]],
                'items': Ramps[dataset['ramp']]['legend']
            }
        
        }
        ramp=Ramps[dictChange_key[nas[1]]]
        units='{0} change'.format(units_key[nas[1]])
            
            
        compareWith={
        'name':'Difference {0}'.format(name),
        'provider': Provider,
        'application': ['prep'],
        "env": "production",
        "default": False,
        'layerConfig':{
            'colorRamp':ramp['colorized'],
            "indicator": dataset['indicator'],
            "noData": -1,
            "type": "loca",
            "compareWith":True,
            "scenario":scenario,
            "temp_resolution":temp_resolution
            },
        'legendConfig':{
                'type': 'gradient',
                'unit':units,
                'items': ramp['legend']
            }
        
        }
        dataset['layer']=baseLayer
        dataset['layerCompare']=compareWith
        datasetsLayers.append(dataset)
    return datasetsLayers
```

```python
def createLayer(dataset):
    layerUrl='{0}{1}/layer'.format(apiBaseUrl, dataset['datasetId'])
    r=requests.post(layerUrl, data=json.dumps(dataset['layer']), headers=Headers)
    s=requests.post(layerUrl, data=json.dumps(dataset['layerCompare']), headers=Headers)
    
    return {'layerD':r.json()['data']['id'], 
        'layerCom':s.json()['data']['id']}
```

```python
def updateLayer(dataset):
    layerUrl='{0}{1}/layer/{2}'.format(apiBaseUrl, dataset['datasetId'], dataset['layerId']['layerD'])
    layerUrlC='{0}{1}/layer/{2}'.format(apiBaseUrl, dataset['datasetId'], dataset['layerId']['layerCom'])
    
    r=requests.patch(layerUrl, data=json.dumps(dataset['layer']), headers=Headers)
    print(r.status_code)
    if r.status_code!=200:
        print(r.text)
    s=requests.patch(layerUrlC, data=json.dumps(dataset['layerCompare']), headers=Headers)
    print(s.status_code)
    if s.status_code!=200:
        print(s.text)
    return {'layerD':r.json()['data']['id'], 'layerCom':s.json()['data']['id']}
```

```python
def deleteLayers(datasets):
    for dataset in datasets:
        LayersUrl= '{0}{1}/layer'.format(apiBaseUrl,dataset['datasetId'])
        layers=requests.get(LayersUrl).json()['data']
        for layer in layers:
            LayerUrl= '{0}{1}/layer/{2}'.format(apiBaseUrl,dataset['datasetId'],layer['id'])
            r=requests.delete(LayerUrl, headers=Headers)
```
