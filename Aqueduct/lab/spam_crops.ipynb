````python
import requests
from tqdm import tqdm
import pandas as pd
import numpy as np
````

This script will be use to download all csv related SPAM crops. We are only fetching for the csv as we alredy have the geometries.

The next figure will hold all configurations needed to download the data.To do so we will need to compund this url:
``` 
http://spam05.harvestchoice.org/v2r0/csv/<areas>/spam2005v2r0_<areas>_<crop>_<types>.csv.gz
```  
example:
http://spam05.harvestchoice.org/v2r0/csv/physical-area/spam2005v2r0_physical-area_cereals_other_total.csv.gz

So we will need to identify wich variables we need.
* [Areas](http://mapspam.info/methodology/#physical_area) list of available SPAM outputs:   

|name| available for download |description|  
|:---|:---|:---|  
|PHYSICAL AREA| physical-area|Physical area is measured in hectare, and represents the actual area where a crop is grown, not counting how often production was harvested from it. Physical area is calculated for each  production system and crop, and the sum of all physical areas of the four production systems constitute the total physical area for that crop. The sum of the physical areas of all crops in a pixel may not be larger than the pixel size.|   
|HARVESTED AREA|harvested-area|Also measured in hectare, harvested area is at least as large as physical area, but sometimes more, since it also accounts for multiple harvests of a crop on the same plot. Like for physical area, the harvested area is calculated for each production system, and the sum of all harvested areas of all production systems in a pixel amount to the total harvested area of the pixel.|  
|PRODUCTION| production|Production, for each production system and crop, is calculated by multiplying area harvested with yield. It is measured in metric tons. The total production of a crop includes the production of all production systems of that crop.|  
|YIELD| yield|Yield is a measure of productivity, the amount of production per harvested area, and is measured in kilogram/hectare. The total yield of a crop, when considering all production systems, is not the sum of the individual yields, but the weighted average of the 4 yields|  

* [Crops](http://mapspam.info/wp-content/uploads/4-Methodology-Crops-of-SPAM-2005-2015-02-26.csv): list of available crops  

* [types](http://mapspam.info/methodology/#irrigation_areas): list of available production systems

|name| available for download |
|:---|:---|  
|Irrigated|irrigated|  
|Rainfed|rainfed|   
|Total|total|   


````python
baseUrl = 'http://spam05.harvestchoice.org/v2r0/csv/'

filename = 'harvested-area_banana_total.csv.gz'
areas = ['physical-area']
crops = ['cereals_other', 'pulses_other', 'banana', 'barley', 'bean', 'cassava', 'chickpea', 'cowpea', 'groundnut', 
         'lentil', 'maize', 'millet_pearl', 'millet_small', 'pigeonpea', 'plantain', 'potato', 'rice', 'sorghum', 'soybean', 'sweet_potato',
         'wheat', 'yam']  
types = ['irrigated','rainfed']
````

````python
for typez in xrange(0,len(types)):
    for area in xrange(0,len(areas)):
        for crop in xrange(0,len(crops)):
            dataUrl=baseUrl+areas[area]+'/'+'spam2005v2r0'+'_'+areas[area]+'_'+crops[crop]+'_'+types[typez]+'.csv.gz'
            response = requests.get(dataUrl, stream=True)
            filename=areas[area]+'_'+crops[crop]+'_'+types[typez]+'.csv.gz'
            print filename
            with open(filename, 'wb') as handle:
                for chunk in tqdm(response.iter_content(chunk_size=128)):
                    handle.write(chunk)
````

##### After downloading the data we will unzip, and merge them together into one file.

````python
! gunzip *.gz #command line to uncompress and deleted compresed files
````

````python
baseUrl = './physical-area'
## This will loop over all csv files and will extract the information required to merge them together into total_crops.csv.
for typez in xrange(0,len(types)):
    for crop in xrange(0,len(crops)):
        dataUrl=baseUrl+'_'+crops[crop]+'_'+types[typez]+'.csv'
        print dataUrl
        a=pd.read_csv(dataUrl, sep=',')
        a['irrigation'] = np.where(a['cell5m']>0, types[typez], types[typez])
        a['crop'] = np.where(a['cell5m']>0, crops[crop], crops[crop])
        a.columns.values[4] = 'value'
        a.to_csv('./'+baseUrl+'_'+crops[crop]+'_'+types[typez]+'.csv1', sep=',')

!cat *.csv1 > total_crops.csv2          #command line
!rm *.csv1                              #command line
!rm *.csv                               #command line
!cp total_crops.csv2 total_crops.csv    #command line
!rm *.csv2                              #command line
````
