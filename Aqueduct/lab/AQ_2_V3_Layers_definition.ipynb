Overall Water Risk (100%)

`Sql:`
```sql
SELECT s.aq30_id as cartodb_id, coalesce(NULLIF(gtd_label,''), 'No Data') as label, r.the_geom, r.the_geom_webmercator,
(CASE WHEN gtd_label = 'Insignificant Trend' THEN -1 ELSE coalesce(gtd_cat, -9999)END) as water_risk
FROM water_risk_indicators_annual s 
LEFT JOIN y2018m12d06_rh_master_shape_v01 r on s.aq30_id=r.aq30_id 
WHERE s.pfaf_id != -9999 and s.gid_1 != '-9999' and r.aqid != -9999 
ORDER BY s.aq30_id
```

Weight fraction is the fraction [0-1] of valid data. 1 means 100% of the weighted 13 indicators are valid. 

For example 0.23 means that 23% of the weighted indicators are valid.

We would like to mask out geometries <0.75 (less than 75% valid data).

`Sql:`
```sql
with r as (SELECT s.aq30_id as cartodb_id, s.aq30_id, coalesce(w_awr_def_tot_cat, -9999) as  water_risk,  
           coalesce(NULLIF(w_awr_def_tot_label,''), 'NoData') as label, w_awr_def_tot_weight_fraction as fraction,
           r.the_geom_webmercator, r.the_geom
		   FROM water_risk_indicators_annual s 
		   LEFT JOIN y2018m12d06_rh_master_shape_v01 r on s.aq30_id=r.aq30_id 
		   WHERE w_awr_def_tot_cat is not null and s.pfaf_id != -9999 and s.gid_1 != '-9999' and r.aqid != -9999)
           
SELECT cartodb_id, aq30_id, r.the_geom_webmercator, r.the_geom,
	   CASE WHEN water_risk = 4 and  fraction > 0.75 THEN 4
			WHEN water_risk = 3 and  fraction > 0.75 THEN 3
			WHEN water_risk = 2 and  fraction > 0.75 THEN 2
			WHEN water_risk = 1 and  fraction > 0.75 THEN 1
			WHEN water_risk = 0 and  fraction > 0.75 THEN 0
			WHEN water_risk = -1 and  fraction > 0.75 THEN -1
            ELSE  -9999
	   END as water_risk,
	   CASE WHEN label = 'Extremely High (4-5)' and  fraction > 0.75 THEN 'Extremely High (4-5)'
			WHEN label = 'High (3-4)' and  fraction > 0.75 THEN 'High (3-4)'
			WHEN label = 'Medium - High (2-3)' and  fraction > 0.75 THEN 'Medium - High (2-3)'
			WHEN label = 'Low - Medium (1-2)' and  fraction > 0.75 THEN 'Low - Medium (1-2)'
			WHEN label = 'Low (0-1)' and  fraction > 0.75 THEN 'Low (0-1)'
			WHEN label = 'No Data' and  fraction > 0.75 THEN 'No Data'
            ELSE  'No Data'
	   END as label
FROM r
```

### Weight distribution for each industry

We have 13 indicators. Grouped in 3 categories. Which weight range values are: 

**[0.25, 0.5, 1, 2, 4]**   

We build a stacked bar widget to show the contribution of each group.

````python
carto_acc='wri-rw'
sql_endpoint='https://{0}.carto.com/api/v2/sql'.format(carto_acc)

query = """
with r as (SELECT industry_short, group_short, round(sum(weight_fraction*100)::numeric, 1) as fractions 
           FROM water_risk_industry_weights 
           GROUP BY industry_short, group_short 
           ORDER BY industry_short, group_short) 
           
SELECT industry_short, array_agg(group_short) as groups, array_agg(fractions) as weight_fractions 
FROM r GROUP BY industry_short 
"""

payload = {'q': query}
r = requests.get(sql_endpoint, params=payload)

df= pd.read_json(json.dumps(r.json()['rows']), orient='records')
df
````

````python
industry_fractions = {}
for n, industry in enumerate(df['industry_short'].unique()):
    industry_fractions[industry] = df['weight_fractions'].iloc[n]
industry_fractions
````

````python
# Data
fractions = list(industry_fractions.values())
locations = list(industry_fractions.keys())
categories = list(df['groups'].iloc[0])

values = []
for i in range(len(categories)):
    values.append((np.array(fractions)[:,i]).tolist())

# Plot
plt.figure(figsize=(10,5))

width = 0.35       # the width of the bars

for i in range(len(values)):
    if i == 0:
        plt.barh(locations, values[i], width, label=categories[i], edgecolor ='k')
        past_values = np.array(values[i])
    else:
        plt.barh(locations, values[i], width, left=past_values, label=categories[i], edgecolor ='k')
        past_values = np.array(values[i]) + past_values

plt.title('Title')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.legend()

plt.show()
````

````python
carto_acc='wri-rw'
sql_endpoint='https://{0}.carto.com/api/v2/sql'.format(carto_acc)

query = """
with r as (SELECT industry_short, group_short, round(sum(weight_fraction*100)::numeric, 1) as fractions 
           FROM water_risk_industry_weights 
           GROUP BY industry_short, group_short 
           ORDER BY industry_short, group_short) 
           
SELECT industry_short, array_agg(group_short) as groups, array_agg(fractions) as weight_fractions 
FROM r GROUP BY industry_short 
"""

payload = {'q': query}
r = requests.get(sql_endpoint, params=payload)

df= pd.read_json(json.dumps(r.json()['rows']), orient='records')
df
````
