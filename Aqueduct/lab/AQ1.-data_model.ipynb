## Data management 
* Datasets location: Carto  
* Account: wri-rw  
* Filter tables by tag ```:aqueduct```

#### Tables: 

````python
params={
    'page':1,
    'per_page':30,
    'tags':'aqueduct',
    'types':'table',
    'kind':'geom'
},
data = requests.get('https://resourcewatch.carto.com/u/wri-rw/api/v1/viz/?tag_name=&q=&page=1&type=&exclude_shared=false&per_page=30&tags=aqueduct&shared=yes&locked=null&only_liked=null&kind=geom&order=updated_at&types=table%2Cremote&deepInsights=false').json()['visualizations']
t=[{'table_name':x['table']['name'].split('.')[1], 'url':x['url'], 'size':x['table']['size'], 'row_count':x['table']['row_count']}  for x in data if x['kind']=='geom']
table = pd.DataFrame(t)
````
