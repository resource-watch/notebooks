``````python
def find_between(s, first, last):
    try:
        start = s.index(first) + len(first)
        end = s.index(last, start)
        return s[start:end]
    except ValueError:
        return ""
    
def parseStr(key, value):
    if key == 'line-dasharray':
        return [int(i) for i in value.split(",")]
    else:
        try:
            return float(value)
        except:
            return value
#Fix cartocss parser:
def cartoToVectorStyles(css):
    
    """
    Function to parse the cartocss style to the mapbox style. Requires cartocss and returns the parsed vector style.
    
    """
    parsed = []
    for item in css.split('}'):
        tmp_str = item + '}'
        #print(tmp_str)
        tmp_style = find_between(tmp_str,'{', '}')
        #print(f'style: {tmp_style}\n')

        try:
            styles = {s.split(':')[0].strip(): s.split(':')[1] for s in tmp_style.split(';') if s.strip()}
            #print('styles_',styles)
        except:
            print(tmp_style.split(';'))

        for style_type in ['line', 'polygon', 'marker']:
            #print('style_type:', style_type)

            template = {
            'paint': {},
            'source-layer': 'layer0',
            'type': ''
            }

            type_dict={
                    'polygon':'fill',

                    'line': 'line',
                    'marker': 'circle'

                }
            style_dict = {
                'polygon':{
                    'polygon-fill': 'fill-color',
                    'polygon-opacity': 'fill-opacity'
                },
                'line':{
                    'line-color': 'line-color',
                    'line-width': 'line-width',
                    'line-opacity': 'line-opacity',
                    'line-dasharray': 'line-dasharray'
                },
                'marker':{
                    'marker-width': 'circle-radius',
                    'marker-fill': 'circle-color',
                    'marker-fill-opacity': 'circle-opacity',
                    'marker-line-color': 'circle-stroke-color',
                    'marker-line-width': 'circle-stroke-width',
                    'marker-line-opacity': 'circle-stroke-opacity',
                },
                #'zoom':{
                 #   'zoom':'zoom'}
            }

            cond_dict = {
                '>=': '>=',
                '<=': '<=',
                '=': '==',
                '>': '>',
                '<': '<',
            }

            if type_dict.get(style_type,False) == False:
                template['paint'] = 'Style dictionary not defined'
                #print(template['paint'])

            else:
                try:
                    paint = {style_dict[style_type][k]: parseStr(k,v) for k, v in styles.items() if k in style_dict[style_type]}
                    template['paint'] = paint
                    #print('Paint:',paint)
                except:
                    print(f'{style_type} not in styles')

            if type_dict.get(style_type, False) == False:
                        template['type'] = 'Type dictionary not defined'

            else:
                template['type'] = type_dict.get(style_type, False)

            #print('template:', template)

            conditions = ['all']
            done = False
            #print('ITEM',item)
            if '][' in item:
                x = ",".join(item.split(']['))
                tmp_cond = find_between(x, '[', ']').split(',')

                if type(tmp_cond) == list:
                    for el in tmp_cond:
                        done = False
                        for k,v in cond_dict.items():
                            if k in el and not done:
                                done = True
                                col = el.split(k)[0]
                                cond = el.split(k)[1]
                                #col, cond = el.split(k)

                                conditions.append([v, col, parseStr('',cond.replace("'", "").replace('"', ''))]) 
                               # print('conditions', conditions)
            else: 
                tmp_cond = find_between(tmp_str,'[', ']')

                #print('tmp_cond:', tmp_cond)
                for k,v in cond_dict.items():
                    if k in tmp_cond and not done:
                        #print('k',k)
                        col, cond = tmp_cond.split(k)
                        #print('col', col)
                        #print('cond', cond)
                        done = True

                        if col == 'zoom':
                          ## stop! record id, cannot be fixed!
                            #print('ZOOM')
                            return 'zoom'
                        else:
                            conditions.append([v, col,cond])

                    #conditions.append([v, col, str(parseStr('',cond.replace("'", "").replace('"', '')))])

            template['filter'] = conditions
            #print('TEMPLATE:', template)
            parsed.append(template)

    parsed = [p for p in parsed if p['paint']]
    if parsed: return parsed
    return False


def ParseValidatorStyle(track_error, error_name):
    
    """
    Function to parse the vector style for the error layers to the structure needed to run the Mapbox Validator.
    Requires track error and error name:
        - track error: element with the layer id and the error message
        - error name: name of the error to be exported in the json file
        
        """

    
    layer_id = track_error['id']
    
    validator_template = {
            "version": 8,
            "name": layer_id,
            "metadata": {},
            "sources": {},
            "layers":[]
        }
    
    try:
        layer =lmi.Layer(layer_id)
        vlconfig =layer.attributes['layerConfig']['body']['vectorLayers']

        temp_sources = {}
        temp_layers = []
        for i in range(0,len(vlconfig)):
            temp_sources[f'source_{i}'] = {"type": "vector"}

            vl = vlconfig[i]
            try:
                vl["id"] = f"id_{i}",
                vl["source"] = f"source_{i}"
                temp_layers.append(vl)
            except:
                print(layer_id)

        validator_template['sources']=temp_sources
        validator_template['layers']=temp_layers

        #export geojson locally
        with open(f'./{error_name}_toValidate_{layer_id}.json','w') as f:
            json.dump(validator_template,f)
        
        
    except:
        print(f'ERROR: layer {layer_id} does not exist')
        
def ValidateMapbox(list_toValidate):
    """
    Function that validates the mapbox style with the Mapbox Validator. Requires list of layers to validate
    """
    validator = {}
    for el in list_toValidate:
        layer_id = el.split('_')[2].split('.')[0]
        print('layer_id:', layer_id)
        message = !gl-style-validate {el}
        print('message:',message)
        validator[layer_id]=message
    return validator
``````

``````python
#run the css2mapbox script over the layers:
def rewriteVLStyle(error_list,folder_name):
    
    """Function that selects the id of the layers for each error list and generates a new vector style using the updated
    carto2mapbox script
    
    """
    for el in error_list:
        layer_id = el['id']
        try:
            layer = lmi.Layer(layer_id)

        except:
            print(f'Layer {layer_id} does not exist')


        files = os.listdir('./error_0_vlTemplate/')
        if not f"{layer_id}.json" in files:
            flag = False
            lc = layer.attributes['layerConfig']
            print('LayerConfig:',lc)

            try:
                css = layer.attributes['layerConfig']['body']['layers'][0]['options']['cartocss']
                print('css:', css)
                if type(css)== str and len(css)>0:
                    flag = flag
                    vector_templ= cartoToVectorStyles(css)
                    print('vector templ:',vector_templ)
            except:
                flag = True
                vector_templ = 'Vector template not defined'

            tmp_structure = {

            'id': layer_id,
            'css': css,
            'vector': vector_templ,
            'flag': flag
            }
            
            if os.path.isdir(f'./{folder_name}'):
                             pass
            else:
                os.mkdir(f'./{folder_name}')

            with open(f'./{folder_name}/' + f"{layer_id}.json", 'w') as fp:
                json.dump(tmp_structure, fp)

        else:
            print('json already exist')
            
            
def vectorStyleGenerator(path):
    """
    Function that checks the number of layers in a given path with vectorstyle configuration
    """
    #check how many layers we have with no vector style after running the script
    no_vector_true   = []
    toValidate = []
    for el in os.listdir(path):
        if el == '.ipynb_checkpoints':
            pass
        else:
            with open(path+el,'r') as f:
                json_temp = json.load(f)
            flag = json_temp['flag']
            if flag == True:
                layer_id = json_temp['id']
                no_vector_true.append(layer_id)
            else:
                toValidate.append(json_temp) 
    print(f'There are {len(no_vector_true)} layers with no vector style')
    print(f'There are {len(toValidate)} layers with new Vector style to be validated')
    return toValidate
    
    
def NewParseValidatorStyle (list2Validate,error_name):
    for el in list2Validate:
        layer_id = el['id']

        validator_template = {
            "version": 8,
            "name": layer_id,
            "metadata": {},
            "sources": {},
            "layers":[]
        }

        vlconfig = el['vector']

        temp_sources = {}
        temp_layers = []
        for i in range(0,len(vlconfig)):
            temp_sources[f'source_{i}'] = {"type": "vector"}

            vl = vlconfig[i]
            try:
                vl["id"] = f"id_{i}",
                vl["source"] = f"source_{i}"
                temp_layers.append(vl)
            except:
                print(layer_id)

        validator_template['sources']=temp_sources
        validator_template['layers']=temp_layers

        #export geojson locally
        with open(f'./{error_name}_toValidate_{layer_id}.json','w') as f:
            json.dump(validator_template,f)
        
def UniqueErrors(validator):
    list_errors = []
    for k,v in validator.items():
        for el in v:
            try:
                error = el.split('id:')[1]
            except:
                error = el
            list_errors.append(error)
    return set(list_errors)
    
``````
